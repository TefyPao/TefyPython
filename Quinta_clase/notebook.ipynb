```python
##Manejo de strings
## un string es una composición de caracteres
## es posible la indexación, contando desde cero.
## es posible hacer slicing de strings
## un string es inmutable 


string_1= "Hola, ¿Cómo estás?"
```

```python
string_1[3]
```

```python
string_1[1:4],
string_1[4:], # desde la posición 4 hasta  el final 
string_1[:4], # comienza desde cero hasta la posición 4
string_1[1::2] ## desde la posición 1 hasta el final con paso 2
string_1[-1::-1] #para invertir un string 
string_1[::-1] #para invertir

```

```python

string_2= string_1.replace("Cómo", "Dónde") # puede modificar cualquier string o substring
string_2
```
