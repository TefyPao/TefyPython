# Funciones y métodos 
## las funciones retornan un valor, mientras que los métodos, no 


```python
def calcular_perimetro_rectangulo(largo, ancho):
    perimetro= 2*largo + 2*ancho
    return perimetro 
```

```python
p= calcular_perimetro_rectangulo(5,3)
print("El perímetro del rectángulo es: ", p)
```
